# **Training Model**

```python
df.columns
```

```python
X = df[["Avg. Session Length", "Time on App", "Length of Membership"]]
y = df["Yearly Amount Spent"]
```

```python
#import train_test_split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 42)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
li = LinearRegression()
```

```python
li.fit(X_train, y_train)
```

```python
li.coef_
```

```python
co = pd.DataFrame(li.coef_, X.columns, columns=['coef'])
co
```

```python
#time to predict
y_preds = li.predict(X_test)
y_preds
```

```python
sns.regplot(x=y_preds, y=y_test)
plt.xlabel("Yearly Amount Spent Predicitons")
plt.title("Comparing the fitting of the Yearly Amount Spent")
```

```python
#residuals
residuals = y_test - y_preds
residuals
```

```python
sns.displot(residuals, bins = 20, kde = True)
```

```python
#import metrics
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
print("R2_score:", r2_score(y_preds, y_test)),
print("Mean_absolute_error:", mean_absolute_error(y_preds, y_test))
print("Mean_squared_error:", mean_squared_error(y_preds, y_test))
```
